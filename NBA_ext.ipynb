```python
from nba_api.stats.endpoints import leaguedashplayerstats
from nba_api.stats.endpoints import playercareerstats
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)       # Show all rows
pd.set_option('display.max_columns', None)    # Show all columns
pd.set_option('display.width', 1000)          # Set the width to accommodate more columns
pd.set_option('display.max_colwidth', None)   # Set maximum column width to None (auto adjust)
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python


# Define the seasons for the Jordan and LeBron eras
jordan_era_seasons = [
    '1984-85', '1985-86', '1986-87', '1987-88', '1988-89', '1989-90', '1990-91', '1991-92', '1992-93',
    '1994-95', '1995-96', '1996-97', '1997-98'
]
lebron_era_seasons = [
    '2003-04', '2004-05', '2005-06', '2006-07', '2007-08', '2008-09', '2009-10', '2010-11', '2011-12', 
    '2012-13', '2013-14', '2014-15', '2015-16', '2016-17', '2017-18', '2018-19', '2019-20', '2020-21', 
    '2021-22', '2022-23'
]

# Initialize empty lists to store data
jordan_era_data = []
lebron_era_data = []

# Fetch league-wide stats for Jordan Era
for season in jordan_era_seasons:
    stats = leaguedashplayerstats.LeagueDashPlayerStats(season=season)
    stats_df = stats.get_data_frames()[0]
    stats_df['SEASON'] = season
    jordan_era_data.append(stats_df)

# Fetch league-wide stats for LeBron Era
for season in lebron_era_seasons:
    stats = leaguedashplayerstats.LeagueDashPlayerStats(season=season)
    stats_df = stats.get_data_frames()[0]
    stats_df['SEASON'] = season
    lebron_era_data.append(stats_df)

# Combine each era's data into single DataFrames
jordan_era_stats_df = pd.concat(jordan_era_data, ignore_index=True)
lebron_era_stats_df = pd.concat(lebron_era_data, ignore_index=True)

# Display first few rows of each era's data
jordan_era_stats_df.head(), lebron_era_stats_df.head()
```

```python
# Lista de columnas a eliminar
columns_to_drop = [
    'NBA_FANTASY_PTS', 'DD2', 'TD3', 'WNBA_FANTASY_PTS', 'GP_RANK', 'W_RANK', 'L_RANK', 'W_PCT_RANK',
    'MIN_RANK', 'FGM_RANK', 'FGA_RANK', 'FG_PCT_RANK', 'FG3M_RANK', 'FG3A_RANK', 'FG3_PCT_RANK',
    'FTM_RANK', 'FTA_RANK', 'FT_PCT_RANK', 'OREB_RANK', 'DREB_RANK', 'REB_RANK', 'AST_RANK', 
    'TOV_RANK', 'STL_RANK', 'BLK_RANK', 'BLKA_RANK', 'PF_RANK', 'PFD_RANK', 'PTS_RANK', 
    'PLUS_MINUS_RANK', 'NBA_FANTASY_PTS_RANK', 'DD2_RANK', 'TD3_RANK', 'WNBA_FANTASY_PTS_RANK'
]

# Elimina las columnas del DataFrame (ejemplo con jordan_era_stats_df)
jordan_era_stats_df = jordan_era_stats_df.drop(columns=columns_to_drop, axis=1)
lebron_era_stats_df = lebron_era_stats_df.drop(columns=columns_to_drop, axis=1)

# Muestra los primeros registros para verificar que las columnas han sido eliminadas
jordan_era_stats_df.head(), lebron_era_stats_df.head()
```
